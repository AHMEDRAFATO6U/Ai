### svm_linear_model

```python
from sklearn.svm import SVC ## call model 
linear_svm=SVC(kernel='linear' ,C=100)
```

```python
linear_svm.fit(x_train,y_train)
```

```python
x_train=np.array(x_train)
y_train=np.array(y_train)
plot_decision_regions(x_train,y_train ,clf=linear_svm)
plt.xlabel('Feature 1')
plt.ylabel("Feature 2")
plt.title("Linear SVM Decision Boundary")
plt.show()
```

### SVM POLY MODEL

```python
### poly svm
poly_svm=SVC(kernel='poly',degree=3,C=100)
poly_svm.fit(x_train,y_train)
```

```python
x_train=np.array(x_train)
y_train=np.array(y_train)
plot_decision_regions(x_train,y_train ,clf=poly_svm)
plt.xlabel('Feature 1')
plt.ylabel("Feature 2")
plt.title("Linear SVM Decision Boundary")
plt.show()
```

### SVM RBF MODEL

```python
### rbf model
rbf_svm=SVC(kernel='rbf',C=100)
rbf_svm.fit(x_train,y_train)
```

```python
x_train=np.array(x_train)
y_train=np.array(y_train)
plot_decision_regions(x_train,y_train ,clf=rbf_svm)
plt.xlabel('Feature 1')
plt.ylabel("Feature 2")
plt.title("Linear SVM Decision Boundary")
plt.show()
```

### SIGMOID SVM MODEL

```python
# segmoid 
sigmoid_svm=SVC(kernel='sigmoid')
sigmoid_svm.fit(x_train,y_train)
```
